```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
x = np.linspace(-2.0*np.pi, 2.0*np.pi, 101)
y1 = 3*np.cos(x - np.pi/2) + 1
y2 = 0.5*np.cos(x - np.pi/2) + 0.5

xnumbers = np.linspace(-4, 4, 11)
ynumbers = np.linspace(-6, 6, 11)

plt.plot(x, y1, 'y', x, y2, 'r')
plt.xlabel("Angle in Radians")
plt.ylabel("Magnitude")
plt.title("Plot of two trigonometric functions")
plt.xticks(xnumbers)
plt.yticks(ynumbers)
plt.legend(['cosine1', 'cosine2'])
plt.grid()
plt.axis([-5, 5, -4, 5])
plt.show()

```

```python
# 3.1. Напишите код, который будет переводить полярные координаты в декартовы.
import math

def pol2cart(rho, phi):
    x = rho * math.cos(math.radians(phi))
    y = rho * math.sin(math.radians(phi))
    return(x, y)

rho=float(10)
phi=float(1/2)

print("For the polar coordinates x = ", rho, "and y = ", phi, "the cartesian coordinates are = ", pol2cart(rho, phi))
```
