```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 3 (окружность)
circle = plt.Circle((0, 0), 1, color='y', fill=False)

xnumbers = np.linspace(-2, 2, 5)
ynumbers = np.linspace(-2, 2, 5)

ax=plt.gca()
ax.add_patch(circle)
plt.axis('scaled')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Circle graph")
plt.xticks(xnumbers)
plt.yticks(ynumbers)
plt.grid()
plt.show()
```

```python
# 3 (окружность)

t = np.arange(0, 2*np.pi, 0.01)
r = 4

plt.plot(r*np.sin(t), r*np.cos(t), 'y', lw=3)
plt.axis('equal')

plt.grid()
plt.axis([-5., 5., -5., 5.])
plt.show()
```

```python
# 3 (эллипс)

t = np.linspace(0, 2*pi, 100)
plt.plot(1+2*np.cos(t) , 1.5*np.sin(t))
plt.grid()
plt.show()
```

```python
# 3 (гипербола)
import decimal

xmin = -5
xmax = 5
dx = 0.05

xlist = np.around(np.arange(xmin, xmax, dx), decimals=4)

ylist = 1 / xlist

xnumbers = np.linspace(-5, 5, 11)
ynumbers = np.linspace(-5, 5, 11)

plt.axis('scaled')
plt.plot(xlist, ylist, 'y')

plt.xticks(xnumbers)
plt.yticks(ynumbers)
plt.axis([-5.5, 5.5, -5.5, 5.5])
plt.grid()
plt.show()
```

```python
# 5.1
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

X, Y = np.meshgrid(range(3), range(3))

Z = (9 - X - Y) / 2
Z1 = (9 - (X+1) - (Y+1)) / 2

ax.plot_surface(X, Y, Z, alpha=0.5)
ax.plot_surface(X, Y, Z1, alpha=0.5)
plt.show()
```

```python
# 5.1
f = lambda x, y: x ** 2 - y ** 2

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')

xnumbers = np.linspace(-5, 5, 100)
ynumbers = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(xnumbers, ynumbers)
Z = f(X, Y)
Z1 = X**2 + Y**2

ax.plot_surface(X, Y, Z, rstride = 5, cstride = 5, alpha=0.5)
ax.plot_surface(X, Y, Z1, alpha=0.5)
plt.show()
```
