```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1.0*np.pi, 1.0*np.pi, 101)
y1 = np.cos(x)
y2 = np.cos(2*x)
```

```python
xnumbers = np.linspace(-4, 4, 11)
ynumbers = np.linspace(-1, 1, 11)

plt.plot(x, y1, 'y', x, y2, 'r')
plt.xlabel("Angle in Radians")
plt.ylabel("Magnitude")
plt.title("Plot of two trigonometric functions")
plt.xticks(xnumbers)
plt.yticks(ynumbers)
plt.legend(['cosine1', 'cosine2'])
plt.grid()
plt.axis([-3.5, 3.5, -1.1, 1.1])
plt.show()
```
